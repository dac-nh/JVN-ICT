```
import numpy as np
```

```
links = {
    'webpage-1': ['webpage-2', 'webpage-4', 'webpage-5', 'webpage-6', 'webpage-8', 'webpage-9', 'webpage-10'],
    'webpage-2': ['webpage-5', 'webpage-6'],
    'webpage-3': ['webpage-10'],
    'webpage-4': ['webpage-9'],
    'webpage-5': ['webpage-2', 'webpage-4'],
    'webpage-6': [], # dangling page
    'webpage-7': ['webpage-1', 'webpage-3', 'webpage-4'],
    'webpage-8': ['webpage-1'],
    'webpage-9': ['webpage-1', 'webpage-2', 'webpage-3', 'webpage-8', 'webpage-10'],
    'webpage-10': ['webpage-2', 'webpage-3', 'webpage-8', 'webpage-9'],
}
```

```
website_index = {website: index for index, website in enumerate(links.keys())}
website_index
# {'webpage-10': 3, 'webpage-9': 0, 'webpage-8': 1, 'webpage-1': 2, 'webpage-3': 4, 'webpage-2': 5, 'webpage-5': 6, 'webpage-4': 7, 'webpage-7': 8, 'webpage-6': 9}
```

```
def build_A(links, website_index):
    A = np.zeros((len(website_index), len(website_index)))
    for current_webpage in links.keys():
        if not links[current_webpage]:
            print('Dangling page: {}'.format(current_webpage))
        else:
            for dest_webpage in links.keys():
                if dest_webpage in links[current_webpage]:
                    A[website_index[current_webpage]][website_index[dest_webpage]] = round(1/len(links[current_webpage]), 2)
    return A

def pagerank(A, eps, d):
    d = d
    S = np.ones(len(A))/len(A)
    eps = eps
    
    delta = 1
    while delta > eps:    
        S_new = np.ones(len(A))*(1-d)/len(A) + d*A.T.dot(S)
        delta = abs(S_new - S).sum()
        S = S_new
    
    return S
```

```
A = build_A(links, website_index)
A
```

```
S = pagerank(A, 0.0001, 0.85)
S
```

```
print([item[0] for item in sorted(enumerate(S), key=lambda item: -item[1])])
# [0, 3, 5, 2, 6, 9, 1, 7, 4, 8]
```
