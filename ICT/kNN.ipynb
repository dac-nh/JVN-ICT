```
import numpy as np
import matplotlib as plt
import matplotlib.lines as mlines
import time

from sklearn.datasets.samples_generator import make_blobs
```

```
centers = [(20, 20), (10, 1), (0, 20)]
cluster_std = [5, 5, 5]
X, y = make_blobs(n_samples=500, cluster_std=cluster_std, centers=centers, n_features=3, random_state=1)
```

```
plt.pyplot.scatter(X[y == 0, 0], X[y == 0, 1], color="red", s=40, label="Cluster1", alpha=0.8)
plt.pyplot.scatter(X[y == 1, 0], X[y == 1, 1], color="blue", s=40, label="Cluster2", alpha=0.8)
plt.pyplot.scatter(X[y == 2, 0], X[y == 2, 1], color="green", s=40, label="Cluster3", alpha=0.8)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```
knn = KNeighborsClassifier()
params = {
    'n_neighbors': range(1, 21),
    'weights': ['uniform', 'distance']
}
cfi = GridSearchCV(knn, params, cv = 10)

print(cfi.fit(X, y).score(X,y))
```
